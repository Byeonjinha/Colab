```python
#기본 차트
plt.title('plot2')
plt.plot([10,20,30,40,50,60],[1,3,4,9,10,17])
plt.show()
# x축 설정
```

```python
# 차트 스타일
plt.title('plot3')
plt.plot([10, 20, 30, 40,  50, 60],[1,3,4,9,10,17], 'gs:')
plt.grid(True)
plt.show
```
